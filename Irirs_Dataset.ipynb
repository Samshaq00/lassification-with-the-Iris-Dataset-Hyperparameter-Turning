# Import Libraries and Load Dataset

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score, GridSearchCV

iris = load_iris()
X = iris.data
Y = iris.target
```

# 10-Fold-Cross-Validation for Different K Values

```python
k_range = range(1, 31)
k_scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X, Y, cv=10)
    k_scores.append(scores.mean())
```

# Plot the accuracy vs. K Value

```python
plt.figure(figsize=(10, 5))
plt.plot(k_range, k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
plt.title('K vs. Cross-Validated Accuracy')
plt.show()
```

# Zoom In on Optimal K Value

```python
plt.figure(figsize=(10, 5))
plt.plot(k_range[10:16], k_scores[10:16])  # Adjust the range as needed
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
plt.title('Zoomed-In Plot of K vs. Cross-Validated Accuracy')
plt.show()
```

# Confirm Using GridSearchCV

```python
param_grid = {'n_neighbors': range(1, 31)}
grid = GridSearchCV(KNeighborsClassifier(), param_grid, cv=10)
grid.fit(X, Y)

print(f"Best parameter (k): {grid.best_params_['n_neighbors']}")
print(f"Best cross-validated accuracy: {grid.best_score_}")
```
